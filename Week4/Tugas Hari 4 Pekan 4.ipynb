```python
# Read data
import pandas as pd

df = pd.read_csv('titanic.csv')
df.head()
```

```python
df[['PassengerId','Survived','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']].isnull().sum()
```

```python
#Menghapus data unique
df.drop(['Name','Ticket'], inplace = True, axis = 1)
```

```python
#tampilan data yang kita miliki sekarang
df.head()
```

```python
#membersihkan data tidak clean pada sex dan embarked
#mencari nilai unique pada masing-masing kolom
print(df['Sex'].unique())
print(df['Embarked'].unique())
```

```python
obj_sex = {'male' : 0, 'female' : 1}

obj_embarked = {'S': 0, 'C': 1, 'Q': 2}

df['Sex'].replace(obj_sex, inplace = True)
df['Embarked'].replace(obj_embarked, inplace = True)
```

```python
#tampilan data yang kita miliki sekarang
df.head()
```

```python
#membersihkan data tidak clean pada age, cabin, dan embarked
#mengetahui shape data
df.shape
```

Terdapat 891 baris, diketahui dari data tadi bahwa missing values dari kolom Age = 177, Cabin = 687, Embarked = 2, Sehingga tidak mungkin melakukan drop pada kolom umur karena jumlahnya terlalu banyak dan menurut saya kolom dan baris tentang umur penting dalam melakukan pengolahan data. Kolom cabin juga berpengaruh terhadap pengolahan data. Sehingga saya memutuskan menggunakan KNN karena kemungkinan terdapat korelasi antara kolom tersebut

```python
#mengubah data NaN pada cabin menjadi '0'
import numpy as np

df['Cabin'].replace(np.nan, '0', inplace = True)

#mencari nilai unique pada kolom cabin
print(df['Cabin'].unique())
```

```python
#mengubah data kolom cabing dari string menjadi float
k = df['Cabin'].unique()
v = np.arange(0, len(k))

obj_cabin = dict(zip(k,v))

df['Cabin'] = df['Cabin'].replace(obj_cabin)
df['Cabin'] = df['Cabin'].replace(0, np.nan)
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors = 3)
miss_value = ['Age','Cabin','Embarked']
df[miss_value] = imputer.fit_transform(df[miss_value])

df
```

```python
# mengetahui jumlah orang yang selamat
df['Survived'].value_counts()
```

```python
#menentukan model dari data, disini saya menggunakan KNeighborsClassifier 
from sklearn.neighbors import KNeighborsClassifier

X = df.drop('Survived', axis = 1)
y = df['Survived']
```

```python
#preprocessing data
from sklearn.preprocessing import StandardScaler

stds = StandardScaler()
datastd = stds.fit_transform(X)
X = pd.DataFrame(datastd, columns = X.columns)

X
```

```python
#menentukan hyperparameter
from sklearn.model_selection import GridSearchCV

model = KNeighborsClassifier()
param_grid = {
              'n_neighbors' : np.arange(1,50), 
              'weights': ['distance','uniform'],
              'algorithm' : ['auto','ball_tree','kd_tree','brute']
             }

gscv = GridSearchCV(model, param_grid, scoring = 'accuracy', cv = 5)
gscv.fit(X,y)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```
