Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
import matplotlib.pyplot as plt
# code here
fig, ax = plt.subplots(figsize = (9,7))
ax.plot(x, cos, color = 'r', label = 'Cosx')
ax.plot(x, cos3, color = 'g', label = '3 Cosx', linestyle = '--')
ax.plot(x, sin, color = 'b', label = 'Sin(x-90)', linestyle = '-.')
ax.plot(x, sin2, color = 'y', label = '2 Sin x', marker = 'x', linestyle = '')

ax.set_title('Grafik Cos Sin')
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.legend(loc = 'center')

plt.show()
```
